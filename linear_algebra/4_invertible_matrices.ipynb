## Invertible Matrix Definition

### Definition

**Invertible/Non-Singular**<br/>
A square matrix $A = (a_{ij}) \in \M^{n\times n}$ is *invertible* or *nonsingular* <br/>
> $A$ is invertible $\iff \exists B \in \M^{n\times n} : AB = BA = I$ <br/>


### Matrix Properties

> ** $B$ is unique**<br/>
> > $AB = BA = I \land AC = CA = I$, <br/>
> > then $B = BI = B(AC) = (BA)C = IC = C$ <br/>
>
> **Inverse of $A$** is *denoted* by $A^{-1}$
>
> ** $A$ is Non-singular** means $A$ has an inverse
> > $AA^{-1} = A^{-1}A = I$ <br/>
> > then, $(A^{-1})^{-1} = A$ <br/>
> > *therefore* $B = A^{-1} \iff A = B^{-1}$
>
> **Singular Matrices** have no matching inverse 
> > $\not\exists B \in \M^{n\times n} : AB = BA = I$

**Proving**<br/>
Proving a Matrix is invertible
> $A = (a_{ij}), B = (b_{ij}) \in \M^{n\times n} : AB = I \land BA = I$ <br/>
> $A, B$ are inverse of each other <br/>
>
> To  *prove* that a matrix is *invertible*, only one of the following equalities need to be checked.
> - $AB = I$<br/>
> - $BA = I$

**Determinants must not be 0** <br/>
$\exists A^{-1} \iff |A| \neq 0$ <br/>
*Proof:*
> $A,B \in \M^{n\times n}$ <br/>
> $A = \begin{pmatrix}a &b \\ c &d \\ \end{pmatrix}, B = \begin{pmatrix}x_1 &x_2 \\ y_1 &y_2 \\ \end{pmatrix}$ <br/>
>
> *As $A$ and $B$ are inverses, $AB = I$* <br/>
> > $ax_1 + by_1 = 1, ax_2 + by_2 = 0, cx_1 + dy_1 = 0, cx_2 + dy_2 = 1$ <br/>
>
> If $|A| = ad - bc \neq 0$, systems can be solved by <br/>
> > $x_1 = \frac{d}{|A|}, y_1 = - \frac{c}{|A|}, x_2 = - \frac{b}{|A|}, y_2 = \frac{a}{|A|}$ <br/>
>
> $\msi B = \begin{pmatrix}\frac{d}{|A|} &-\frac{b}{|A|} \\ -\frac{c}{|A|} &\frac{a}{|A|} \\ \end{pmatrix} = \frac{1}{|A|} \begin{pmatrix} d &-b \\ -c &a \end{pmatrix}$

### Products

**Non-singular matrix products are non-singular** <br/>
$A^{-1}B^{-1} = (AB)^{-1}$
> $I = AA^{-1} = A(I)A^{-1} = A(BB^{-1})A^{-1} = (AB)(B^{-1}A^{-1})$ <br/>
>
> **Determinant of $A^{-1}$ is the reciprocal of the determinant of $A$**<br/>
> $|A^{-1}| = |A|^{-1}$
> > as $AA^{-1} = I$, $|AA^{-1}| = |A||A^{-1}| = 1$
