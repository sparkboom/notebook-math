## Measures of Central Tendency

Typically to summarise a data set, start with a measure of central tendency

### Mean (Arithmetic)
*The sum of all values, divded by the number of data points* <br/>
$\displaystyle AM(X) = GnM_1(X) = \frac{1}{n}\cdot\sum_{i=1}^{n} a_i$
> 
> Appropriate for
> - Data with a normal distribution (outliers can skew results)
> - Continuous equal-interval data
> 
> Examples
> - Half-marathon finish times (no professional runners in set)
> 
> Domain/Range
> - Domain must be non-empty
> - Range is valued
> - Range isnNot unique


#### Pythagorean Means
The Arithmetic, Geometric, and Harmonic means are known collectively as the Pythagorean means<br/>
- Known Inequality: $AM \ge GM \ge HM$
- $HM = \frac{1}{AM}$
- $GM = \frac{1}{?}$

#### Geometric Mean
$\displaystyle GM(X) = \Big( \prod_{i=1}^{n} a_i \Big)^{\frac{1}{n}}  $ or $ \exp\Big( \int_{\Omega_X} \log(x) .df_x \Big)$
> Appropriate for 
> - Handling measure of things known to grow exponentially 
> - Help estimating geometric means with regression models.
>
> Examples
> - income
> - bacterial colonies
> - disease progression
> - biostatistics
>
> Domain/Range
> - Domain must be positive numbers

#### Harmonic Mean
$\displaystyle HM(X) = GnM_{-1}(X) = \Big( \frac{1}{n} \cdot \sum_{i=1}^n \frac{1}{a_i}   \Big)^{-1}$ or $\Big( \int_{\Omega_X} x^{-1} .df_x \Big)^{-1}$
> Appropriate for 
> - Rate where you have incidents, tasks or events in a numerator & measures of person time in denominator
>
> Examples
> - Health Planning
> - Corporate Mergers
> - Speed of fixed-distance trips

#### Quadratic Mean/RMS
$\displaystyle QM(X) = GnM_2(X) = \Big( \frac{1}{n} \cdot \sum_{i=1}^{n} x_i^2 \Big)^{\frac{1}{2}}$

#### Cubic Mean
$\displaystyle CM(X) = GnM_3(X) = \Big( \frac{1}{n} \cdot \sum_{i=1}^{n} x_i^3 \Big)^{\frac{1}{3}}$

#### Generalized Mean
$\displaystyle GnM_p(X) = \Big( \frac{1}{n} \cdot \sum_{i=1}^{n} x_i^p \Big)^{\frac{1}{p}}$

#### Weighted Mean
$\displaystyle \frac{ \sum\limits_{i=1}^{n} \omega_i x_i }{ \sum\limits_{i=1}^{n} \omega_i }$

### Median
*The middle (or average of middle 2 if even number of data points), value of a set*<br/>
*(For a continuous system, it can be considered the place where values are equally likely to appear on the left side as the right. )*
> Appropriate for 
> - Data with outliers (Not likely to be distorted by extremely high/low values)
> - looking for a 'Typical' value, rather an central point
> - Ordered data (necessary)
>
> Examples
> - Family income level
> - Life expectancy in middle ages (mean is in 30's, high death mortality in children pulled this value down; adults often lived to 50's and above )
> - Calculating average number of feet
>
> Range
> - Valued or NaN if data set empty <br/>
> - Not unique

### Mode
*The most common value in a set*
> Appropriate for 
> - nominal data (non-numeric values) 
> - relatively large sampe
> - bi/multimodal data
> 
> Examples
> - shoe size, 
> - most common first names or surnames, 
> - voting systems, such as amazon stars 
> - favourite colour 
>
> Range 
> - can be infinite, undefined, or valued <br/>
> - Always unique<br/>
> - Can have multiple values - bimodal (2)/multimodal

## Skew

If Mean and Median are the same, the data is symmetric - there's equal amount of data on either side of the bell curve.  It has 0 skew

Skewed data will show a peak at the Mode, Median will balance area under graph then mean will be further out




```python
np.random.seed(1000) # Fix the seed so figures are consistent when generated

def plot_dist(a, plt, title="Distribution"):
    # percent point function
    x = np.linspace(stats.skewnorm.ppf(0.01, a), stats.skewnorm.ppf(0.99, a), 100)

    # probability density function
    y = stats.skewnorm.pdf(x, a) 

    # random variates - produce random variables that fit this distribution
    r = stats.skewnorm.rvs(a, size=50000)

    # mode works best with nominal discrete values, so we'll distribute the generated values into equal size bins with range 0.1
    r_binned = np.around(r,1)
    x_mode = stats.mode(r_binned)
    x_median = np.median(r)
    x_mean = r.mean()
    
    plt.title(title)
    plt.plot(x, y, color='slateblue', lw=1, alpha=0.6, label='distribution')
    plt.fill_between(x, y, alpha=0.1, color='slateblue')
    plt.axvline(x_mean, color='red', linestyle='dashed', linewidth=1, label="mean")
    plt.axvline(x_median, color='orangered', linestyle='dashed', linewidth=1, label="median")
    plt.axvline(x_mode.mode, color='goldenrod', linestyle='dashed', linewidth=1, label="mode")
    plt.legend()
    
plt.figure(figsize=(20,6))

plt.subplot(1, 3, 1)
plot_dist(0, plt, 'Normal Distribution')

plt.subplot(1, 3, 2)
plot_dist(3, plt, 'Skewed 3 Normal Distribution')

plt.subplot(1, 3, 3)
plot_dist(-8, plt, 'Skewed -8 Normal Distribution')
    
plt.show();
```
