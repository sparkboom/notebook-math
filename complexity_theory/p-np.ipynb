- Polynomial time 
- $S = f(N), S = \Sigma N^{x_i}$, where N is the measure of size of the input, and x is a postive integer
- Proof that it is fast to do, therefore it is fast to check

## NP Problems
Answer can be checked within a reasonable amount of time
- Non-Deterministic Polynomial Time
- i.e. Polynomial Time to check the correct answer
- No proof that it can be performed fast
- P is a subset of NP
