## Definition - Riemann Hypothesis

### Riemann $\zeta$ Function
$\zeta(s) = 1 + \frac1{2^s} + \frac1{3^s} + \dots + \frac1{n^s} + \dots +$ 

$\zeta(s) = \sum\limits_{n=1}^{\infty} \frac1{n^s}$
> $s \in \C$ <br/>

### Reimann Hypothesis

In critical strip $0 < x < 1$, any solution of $\zeta(s) = 0$ is on 'crticial line' $x = \frac12$

----

## Definition - Euler Product

Euler product

$\displaystyle \zeta(s) = \prod\limits_{p\;prime} \frac1{1- \frac1{p^s} }$
> product over primes $p$

> Infinite sum = infinite product equivalent to unique factorization of integers $n$ by primes $p$

----

## Definition - The Todd Function

- T(s) is a weakly analytic $L^2$ function of a complex variable $s \not= 0$
- for real $s > 0$, T increases from $T(1) = 1$ to $T(\pi) = Ж$
> $T(e^{\pi i s}) = e^{Ж i s}$ <br/>
> *so T maps Euler equation to Euler-Hamilton equation*

- T(s) is analytic on compact sets
- T(s) is polynomial on compact convex sets
- T is defined by an infinite iteration of exponentials

Note:
- *step function* - is weakly analytic but not continuous
- *analytic* -  a convergent power series (can be real and/or complex functions)
- *compactness* - is expressed by how finite the set is

### How does this relate to Reimann Hypothesis?

- Proving RH needs powerful new method
- provided by the Todd function developed to explain the fine structure constant $\alpha$:
> $\frac1\alpha = Ж = 137.035 999...$

The Riemann Hypothesis is a bonus!

## Reimann Hypothesis

Why is RH interesting but difficult?
> - Primes exhibit local irregularity but asymptotic regularity
> - Difficult to calculatre the number of primes less than N
> - The Riemann Hypothesis leads to the best answers

----

## Proof

How does T(s) help prove Riemann Hypothesis?

- Proof by contradiction:
> assume $b$ is a zero of $\zeta$ in the critical strip but not on the critical line & derive cantradiction. Define:
> $F(s) = T(1+\zeta(s+b)) -1$ <br/>
> $F(b) = 0$

- convexity of the critical strip implies $F(2s) = 2F(s)$
- analyticity of $F(s)$ at 0 then implies $F(s)$ identically zero
- this implies $\zeta(s)$ is identicdally zero, the required contradiction
