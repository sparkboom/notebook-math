Algebraic Expression
> $\begin{split}
2&x - &y &= &0 \\
-&x + 2&y &= &3 \\
\end{split}$ <br/>

Linear Combination of columns
> $x\left[\begin{smallmatrix}2\\-1\end{smallmatrix}\right] + y\left[\begin{smallmatrix}-1\\2\end{smallmatrix}\right] =
\left[\begin{smallmatrix}0\\3\end{smallmatrix}\right]$ <br/>

Coefficient Matrix ($Ax=b$)
> $\left[\begin{smallmatrix}2 &-1\\-1 &2\end{smallmatrix}\right] \left[\begin{smallmatrix}x \\ y\end{smallmatrix}\right] = \left[\begin{smallmatrix}0\\3\end{smallmatrix}\right]$

Augmented Matrix ($(A|b)$)
> $
\left[\begin{array}{cc|c}
2 &-1 &0\\
-1 &2 &3\\
\end{array}\right]$

```python
# Solve via NumPy

# Solve for Ax = b
a = np.matrix('2 -1; -1 2')
b = np.array([0, 3])
x = np.linalg.solve(a, b)
x
```

```python
# Solve using SymPy - Equation List Form
x, y = symbols('x, y')
linsolve([2*x - y, -1*x + 2*y - 3 ], (x, y))
```

```python
# Solve using SymPy - Augmented Matrix Form
x, y = symbols('x, y')
linsolve(Matrix(([2, -1, 0], [-1, 2, 3])), (x, y))
```

```python
# Solve using SymPy - A*x = b Form
x, y = symbols('x, y')
M = Matrix(((2, -1, 0), (-1, 2, 3)))

A= M[:, :-1]
b = M[:, -1]
linsolve((A, b), (x, y))
```

### Elimination

Process
> $A \in \M^{nxn}$ <br/>
> For each row $i$ from $1,...,(n-1)$ <br/>
> > Find multiplier $m_i$ of diagonal element $a_{ii}$ that will eliminate the proceeding element in the same column $a_{i+1,i}$ <br/>
> - i.e. if $a_{11} =1, a_{21} =3$, then multiplier $m_1 = -3$<br/>
>
> For each element of the row beneath (i+1), a{i+1,j} <br/>
> where $j$ from $1,...,(n-1)$ <br/>
>> add *to* the element $a_{i+1,j}$ the value $m_i$ multiplied with the preceding value in the column $a{ij}$ <br/>
>> The element beneath the diagonal $a_{i+1,i}$ should now be 0.
>
> Continue, until you have a upper triangular matrix
>
> Then, use back-substitution to resolve values.

#### Example
Equation Set
> $
\begin{split}
&x + &2y + &z &= &2 \\
3&x+&8y + &z &= &12 \\
&&4y+&z &= &2\\
\end{split} $

$Ax = b$ Form
>$
\begin{pmatrix}
1 &2 &1  \\
3 &8 &1  \\
0 &4 &1  \\
\end{pmatrix}$

**1. Elimination - pivot @ $a_{11}$ ($A_1$)**

Pivot first row
> $a_{11}$ is the pivot element. <br/>
> We want to make $a_{21} = 0$ to eliminate the first ($x$) column<br/>
> - multiply $a_{11}$ with 3 and subtract it to $a_{21}$ to make 0<br/>
> - repeat subtraction using the multiplier -3, times the value above for each element in the row <br/>
>
> $\ims
\begin{pmatrix}
\fbox{1} &2 &1  \\
3 &8 &1  \\
0 &4 &1  \\
\end{pmatrix} 
\xrightarrow{\times (-3)}{}
\begin{pmatrix}
1 &2 &1  \\
3+(1\cdot-3) &8+(2\cdot-3) &1+(1\cdot-3)  \\
0 &4 &1  \\
\end{pmatrix}
\xrightarrow{}{}
\begin{pmatrix}
1 &2 &1  \\
0 &2 &-2  \\
0 &4 &1  \\
\end{pmatrix}
$

**2. Elimination - pivot @ $a_{21}$**
> $\begin{pmatrix}
1 &2 &1  \\
\fbox{0} &2 &-2  \\
0 &4 &1  \\
\end{pmatrix} $
><br/>
> *$a_{31} = 0$,<br/> no elimination required, continue to next step*

**3. Elimination - pivot @ $a_{22}$ ($A_2$)**
> $ 
\begin{pmatrix}
1 &2 &1  \\
0 &\fbox{2} &-2  \\
0 &4 &1  \\
\end{pmatrix} 
\xrightarrow{\times (-2)}{}
\begin{pmatrix}
1 &2 &1  \\
0 &2 &-2  \\
0 &0 &5  \\
\end{pmatrix}
$

**4. Upper Traingular Matrix**

> We have an Upper Triangular Matrix U

**Elimination Notes**

You cannot pivot on 0.

> *Solution - perform row exchange to make elimination possible (see later about Permutation matrix)*
>
>$
\begin{pmatrix}
\fbox{0} &4 &1  \\

>
> $
\begin{pmatrix}

\end{pmatrix}
\xrightarrow{\times (-3)}{}\xrightarrow{\times (-2)}{}
\begin{pmatrix}

0 &2 &-2  \\
0 &0 &\fbox{0}  \\
\end{pmatrix} $

Use Augmented Matrix

> $\ims 
\left(\begin{array}{ccc|c}

### Matrix Multiplication 

$A \in \M^{m \times n}, B \in \M^{n \times p}, C \in \M^{m \times p}$ <br/>
such that $AB = C$

1. By Element
> $c_{ij} = \sum\limits_{k=1}^n a_{ik}b_{kj}$

2. By Rows
> For each row $i$<br/>
> $\vec{a_i}B = \vec{c_i}$

3. By Columns
> For each column $j$<br/>
> $A\vec{b_j} = \vec{c_j}$

4. Columns of A x Rows of B
> For each column of A/row of B $k$<br/>
> $col(A,k)\cdot row(B,k) = C_k$, where $C_k \in \M^{m \times p}$ <br/>
>
> Sum the product of each k column of A with k row of B <br/>
> 
> $\sum\limits_{k=1}^n \vec{a}_{*k}\vec{b}_{k*}$

5. By Blocks
> $AB = C$<br/>
> $A = \left(\begin{array}
_ A_1 &A_2  \\
A_3 &A_4 \\
\end{array}\right), B = \left(\begin{array}
_ B_1 &B_2  \\
B_3 &B_4 \\
\end{array}\right)$ <br/>
>
> $C = \left(\begin{array}
_ A_1B_1 + A_2B_3 &A_1B_2 + A_2B_4  \\
A_3B_1 + A_4B_3 &A_3B_2 + A_4B_4  \\
\end{array}\right)$

### Matrix Inverses

A Matrix is *invertible* (or *non-singular*) if
- Must be Square
- $A^{-1}A = I$
- $AA^{-1} = I$

#### Singular Matrix (non-invertible)
> *iff* $\mathbf{|A| = 0}$ (the determinant is 0) <br/>
> *$|A| = 0 \implies$ points are on the same line/plane etc (depending on the no. of dimensions)* <br/>
> Given an $A \in \F^{n\times n}, x \in \F^n$<br/> 
> $Ax=0$, where x is solvable and non-zero<br/>

Example: 
> $\ims
\begin{bmatrix}
1 &3 \\
2 &6 \\
\end{bmatrix}$<br/>
>
> *Notice that both rows and columns have the same ratio.*<br/>
>
> Given an $A \in \F^{n\times n}, x \in \F^n$<br/> 
> $Ax=0$, where x is solvable and non-zero<br/>
>
> $\ims
\begin{bmatrix}
1 &3 \\
2 &6 \\
\end{bmatrix}
\cdot
\begin{bmatrix}
-3  \\
1 \\
\end{bmatrix}
=
\begin{bmatrix}

\end{bmatrix}
$

Gauss-Jordan Method (Solve 2 equations at once)

\underbrace{\begin{bmatrix}

\cdot
\underbrace{\begin{bmatrix}

\underbrace{\begin{bmatrix}

> 
> $\ims
\underbrace{\begin{bmatrix}

\end{bmatrix}}_{A}
\cdot
\underbrace{\begin{bmatrix}

d \\
\end{bmatrix}}_{A^{-1}}
=
\underbrace{\begin{bmatrix}

> $E[A|I] = [I|A^{-1}]$ <br/>
>
> $A|I = \left[\begin{array}{cc|cc}
1 &3 &1 &0 \\
2 &7 &0 &1 \\
\end{array}\right]$<br/><br/>
> Perform elimination 1st to 2nd row to remove the 2 $(\times-2)$<br/>

1 &3 &1 &0 \\

> As $E_{12}E_{21}[A|I], E = E_{12}E_{21}$, then $E[A|I] = [I|A^{-1}] $ <br/><br/>
> $A^{-1} = \begin{bmatrix} 7 &-3\\ -2 &1 \end{bmatrix}$

### Permutation Multiplication 

Permutation Matrix
> $A = LU$ does not account for row exchanges <br/>
> Row exchange can be performed by Permutation Matrix $P$ <br/>
> P = I with reordered rows <br/>
> $PA = LU$ *(for any invertible A)* <br/>

No. of different reorderings
> Count of different permutation matrices sized $n$ is $n!$

### Vector Spaces

What is a 'space'?
> 'space' allows me to hold mathematical constructs such as vectors, alongside their supported operations, with the condition that these operations are closed. <br/>
> 
> i.e. Vectors and vector operations that produce the same type of vectors are spaces.

Examples of Vector Space:
- $\R^2$ all 2 dimensional Real vectors, or x/y plane
- $\R^3$ all 3 dimensional Real vectors, or x/y/z plane
- $\R^n$ all column vectors with Real components

Examples of Non-Vector Spaces:
- $\Z^{+^2}$ all 2 dimensional Positive Integers ($\Z$) - this is not closed, as rational numbers are required as vector element values for some calculations.

Examples of what Vector spaces needs
- needs origin (additive identity)
- need to add (operation)
- need to multiply (operation)

#### Subspaces

A subspace
- is a 'space' so must conform to the definition of a space
- can be a set containing all possible constructs of its space, or a subset of constructs that conform to the definition of space.

Examples of Subspaces
- A line in $R^2$ that intercepts the origin. (the subset of vectors on the line can be added/multiplied and still be on that line)

##### Subspaces of $\R^2$

1. All of plane $\R^2$
2. Any line through $\begin{pmatrix}0 \\ 0\end{pmatrix}$
3. Just the zero-vector $\begin{pmatrix}0 \\ 0\end{pmatrix}$

##### Subspaces of $\R^3$

1. All of hyperplane $R^3$
2. Any plane through the origin $\begin{pmatrix}0 \\ 0\end{pmatrix}$
3. Any line through $\begin{pmatrix}0 \\ 0\\ 0 \end{pmatrix}$
4. Just the zero-vector $\begin{pmatrix}0 \\ 0\\ 0 \end{pmatrix}$

Example of subspace of $\R^3$

$A = \begin{pmatrix} 1 &3 \\ 2 &3 \\ 4 &1 \end{pmatrix}$
> Take all linear combinations of $\begin{pmatrix}1 \\ 2 \\ 4\end{pmatrix}$ and $\begin{pmatrix}3 \\ 4 \\ 1\end{pmatrix}$ <br/>
> - The sum of these two vectors $\begin{pmatrix}4 \\ 5 \\ 5\end{pmatrix}$
> - Any scalar multiplication of all of these vectors 
> - Any vector produced from any combination of legal operations performed on these original vector columns (adding vector columns or multiplying by scalars)
