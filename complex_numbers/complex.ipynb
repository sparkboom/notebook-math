## <a name="set"></a>Number Set


**Set of all complex numbers is $\mathbb{C}$**

> Set of **Imaginary** numbers ($\mathbb{I}$) have no Real ($\mathbb{R}$) component. <br/> 
> Set of **Real** numbers ($\mathbb{R}$) have no Imaginary ($\mathbb{I}$) component. <br/>
> Set of **Complex** numbers $\mathbb{C}$ are a superset of both Real ($\mathbb{R}$) and Imaginary ($\mathbb{I}$) sets. <br/>
> **0** belongs to all sets


```python
# Define sets
imaginary_set = {'Imaginary', '0'}
real_set = {'Real', '0'}
complex_set = {'Imaginary', 'Real', '0','Complex'}
labels = ('', '', '')
all_sets = [imaginary_set, real_set, complex_set]

# Create diagram
v3 = venn3(all_sets, set_labels = labels)
v3.get_label_by_id('101').set_text('Imaginary')
v3.get_label_by_id('001').set_text('Complex')
v3.get_label_by_id('011').set_text('Real')
v3.get_label_by_id('111').set_text('0')

# Display
#plt.title("Real, Imaginary and Complex Number Sets as a Venn Diagram")
#plt.show()
```

## <a name="plane"></a>Complex (Argand) Plane

Complex numbers have 2 dimensions - the real component, and imaginary component. 

```python
# Simple argand view fuction that plots 0 to a position on the complex plane
def argand(a):
    for x in range(len(a)):
        plt.plot([0,a[x].real],[0,a[x].imag],'ro-',label='python')
    limit=np.max(np.ceil(np.absolute(a))) # set limits for axis
    
    plt.xlim((-limit,limit))
    plt.ylim((-limit,limit))
    plt.ylabel('Imaginary')
    plt.xlabel('Real')
    plt.show()
```

```python
a = np.arange(5) + 1j*np.arange(6,11)

# Complex Plane
plt.subplot(2,1, 1)
argand(a)

# Polar plot
plt.subplot(2, 1, 2, polar=True)
for x in a:
    plt.polar([0,np.angle(x)],[0,abs(x)],marker='o')
    
plt.show()
```

## <a name="notation"></a>Complex Notation

### Rectangular Notation
$\lf{ z = a + b\mi }$

#### Real Component ( $\operatorname{Re}(z)$ / $\Re(z)$ )
$\lf{ \Re(z) = a }$

#### Imaginary Component ($\operatorname{Im}(z)$ / $\Im(z)$)
$\lf{ \Im(z) = b }$

### Polar Notation:
$\lf{ z\;\angle\;\theta }$

$\sf{ \text{z is the magnitude or modulus} }$<br/>
$\sf{ \theta \text{ is the angle/argument/phase} }$

## <a name="algebra"></a>Complex Algebra
https://byjus.com/complex-number-formula

### Equality of Complex Numbers Formula

> $\lf{ a+b\mi = c + d\mi \iff (a=c) \land (b = d) }$

### Addition

> $\lf{  (a+b\mi) + (c + d\mi) = (a+c) + (b+d)\mi }$

### Subtraction

> $\lf{  (a+b\mi) - (c + d\mi) = (a-c) + (b-d)\mi }$

### Multiplication

> $\lf{  (a+b\mi)d\mi = -bd + ad\mi }$ <br/>
> $\lf{  (a+b\mi) \times (c + d\mi) = (ac-bd) + (ad+bc)\mi }$

### Complex Reciprocal
*Multiplying by the reciprocal, is the same as dividing*

> $\lf{ \frac{1}{z} = \frac{\bar{z}}{|z|^2} }$
>
> $\sf{ \begin{equation} \label{eq1}
\begin{split}
\frac{1}{z} & = \frac{\bar{z}}{z\bar{z}} \\
& = \frac{a - bi}{a^2 +ab\mi - ab\mi - (bi)^2} \\
& = \frac{a - bi}{a^2 + b^2} \\
& = \frac{\bar{z}}{|z|^2} \\
\end{split}
\end{equation} }$

### Division
*To remove a complex number from the denominator, multiple it by the conjugate*

> $\lf{ \frac{a+b\mi}{c+d\mi} \;=\; \frac{a+b\mi}{c+d\mi} \times \frac{c-d\mi}{c-d\mi} \;=\; \frac{ac+bd}{c^2 + d^2} + \frac{bc-ad}{c^2 + d^2}\mi }$

## <a name="conjugate"></a>Complex Conjugate

if $z = a + b\mi$, then it's conjugate is $\lf{ \bar{z} = a - b\mi}$

### Conjugates - Components

> $\lf{ \Re(\bar{z}) = a }$
>
> $\lf{ \Im(\bar{z}) = -b }$

### Multiply with Conjugate
>*Remove Imaginary component with conjugate*
>
>$\lf{ z\bar{z} = |z|^2 }$<br/>
>
>$\sf{ \begin{equation} \label{eq1}
\begin{split}
z\bar{z} & = (a+b\mi)(a-b\mi) \\
& = a^2 + b^2 \\
& = |z|^2
\end{split}
\end{equation} }$

### Real & Imaginary Functions
>
> $\lf{ \Re(z) \equiv \frac{z + \bar{z}}{2} }$
>
> $\lf{ \Im(z) \equiv \frac{z - \bar{z}}{2} }$

### Equality of Sum of Conjugates
> $\lf{ \overline{z + w} = \overline{z} + \overline{w} }$

## <a name="magnitude"></a>Complex Magnitude

$z = a + b\mi$ can be expressed as the length from the origin and the angle from Real axis $|z|\;\angle\;\theta$

### Magnitude/Modulus Component

> $\lf{ |z|^2 = a^2 + b^2}$

### Angle Component

> $\lf{ \theta = \arctan(\frac{b}{a}) }$
>
> $\lf{ z = |z|\;e^{i\theta} }$
>
> $\lf{ z = |z|\;cos(\theta) + \mi sin(\theta) }$

### Triangle Inequality

> $\lf{ |z + w| \leq |z| + |w| }$
>
> $\lf{ |zw| = |z||w| }$

### Multiplication

> $\lf{ |zw| = |z||w| }$
>
> $\lf{ \angle{zw} = \angle{z} + \angle{w} }$

## <a name="exponent"></a>Complex Exponents

### Exponents

> $\lf{  z^n \implies |z|^n }$
>
>$\sf{ \begin{equation} \label{eq1}
\begin{split}
z^n & = |z||z|... \text{n times} \\
& = |z|^n \\
\end{split}
\end{equation} }$

> $\lf{ z^n \implies \angle\;{n\theta}  }$
>
>$\sf{ \begin{equation} \label{eq2}
\begin{split}
\angle{\theta^n} & = \angle{\theta} + \angle{\theta}... \text{n times} \\
& = \angle{n\theta} \\
\end{split}
\end{equation} }$

### Trig Form

> $\lf{  z^N = (a + bi)^N = |z|^N(\;cos(Nθ) + \mi sin(Nθ)\;) }$

### Exponent Form

> $\lf{ z^N = (|z|\;e^{i\theta})^N =|z|^Ne^{iN\theta} }$
