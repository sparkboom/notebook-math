### Limits - Definition
#### Precise Definition
$\displaystyle\lim_{x\to\alpha} f(x) = L$ 
if for every $\epsilon > 0$, there is  $\delta > 0$, such that whenever $0 < |x-\alpha| < \delta$ then $|f(x) - L| < \epsilon$
#### Working Definition
We say $\displaystyle\lim_{x\to\alpha} f(x) = L$ 
if we can make f(x) as close to L as we want by taking x sufficiently close to $\alpha$ (on either side of $\alpha$) without letting x = $\alpha$.
#### Right Hand Limit Definition
$\displaystyle\lim_{x\to\alpha^+} f(x) = L$, 
same definition as above, abd x>$\alpha$
#### Left Hand Limit Definition
$\displaystyle\lim_{x\to\alpha^+} f(x) = L$, 
same definition as above, and x<$\alpha$

### Basic Derivatives
$\ds \begin{equation} \label{eq1}
\begin{split}
(c f)^\prime & = c f^{\prime}(x), c\in\mathbb{R} \ \\
\\
(f \pm g)^\prime & = f^{\prime}(x) \pm g^{\prime}(x) \\
\\
(f g)^\prime & = f^{\prime}g + fg^{\prime} \\
\\
\left(\frac{f}{g}\right)^\prime & = \frac{f^{\prime}g - fg^{\prime}}{g^2} \\
\\
\dv\Big( f\big(g(x)\big)\Big) & = f^{\prime}\big(g(x)\big)\;g^{\prime}(x) \\
\end{split}
\end{equation}$

### Integration by Parts
#### Derivation
$\newcommand{\dv}{\displaystyle \frac{d}{dx}}$
$\newcommand{\ds}{\displaystyle}$
$\ds \begin{equation} \label{eq1}
\begin{split}
y & = u v \\
\frac{dy}{dx} & = \frac{d(uv)}{dx} \\
\frac{d(uv)}{dx} & = u.\frac{d(v)}{dx} + v.\frac{d(u)}{dx} \\
\text{rearrange...} \\
u.\frac{d(v)}{dx} & = \frac{d(uv)}{dx} - v.\frac{d(u)}{dx} \\
\text{integrate both sides} \\
\int u.\frac{dv}{dx} dx & = \int \Big( \frac{d(uv)}{dx} - v.\frac{du}{dx} \Big) dx \\
\int u.\frac{dv}{dx} dx & = \int \frac{d(uv)}{dx} dx - \int v.\frac{du}{dx} dx \\
\text{formula:} \\
\int u.\frac{dv}{dx} dx & = u.v - \int v.\frac{du}{dx} dx \\
\end{split}
\end{equation}$

### Quadratic Formula
#### Derivation
$\newcommand{\dv}{\displaystyle \frac{d}{dx}}$
$\newcommand{\ds}{\displaystyle}$
(from https://www.mathsisfun.com/algebra/completing-square.html)


$\ds \begin{equation} \label{quadratic}
\begin{split}
ax^2 + bx + c & = 0 \\
x^2 + \frac{b}{a}x + \frac{c}{a} & = 0  \\
x^2 + \frac{b}{a}x & = - \frac{c}{a}  \\
\text{complete the square} \\
\Big(x + \frac{b}{2a}\Big)^2 - \Big(\frac{b}{2a}\Big)^2 & = - \frac{c}{a}  \\
\text{complete the square} \\
\Big(x + \frac{b}{2a}\Big)^2 & = - \frac{c}{a} + \Big(\frac{b}{2a}\Big)^2 \\
\Big(x + \frac{b}{2a}\Big)^2 & = - \frac{c}{a} + \frac{b^2}{(2a)^2} \\
\Big(x + \frac{b}{2a}\Big)^2 & = - \frac{4ac}{4a^2} + \frac{b^2}{(2a)^2} \\
\Big(x + \frac{b}{2a}\Big)^2 & = \frac{b^2 - 4ac}{(2a)^2} \\
x + \frac{b}{2a} & = \pm \sqrt{ \frac{b^2 - 4ac}{(2a)^2} } \\
x & = \frac{\pm \sqrt{ b^2 - 4ac}}{2a} - \frac{b}{2a}\\
x & = \frac{-b \pm \sqrt{ b^2 - 4ac}}{2a}\\
\end{split}
\end{equation}$
