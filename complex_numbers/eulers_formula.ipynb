# Euler's Formula

Euler's Formula
> $\ims \forall x \in \R$ (and also $\C$)<br/>
> $i = \sqrt{-1}$<br/>
> $x  \text{ in radians}$ <br/>
>
> &#x2003; $\msi e^{ix} = \cos x + i \sin x$

*cis abbreviation*
> $\msi \begin{split}
\cis x &= e^{ix} 
&= \cos x + i \sin x
\end{split}$ <br/>

Euler's Identity
> $e^{i\pi} +1 = 0$<br/>
> &#x2003; when $x = \pi$

----

## Applications

$e^{i\varphi}$ is a *unit complex number* <br/>
> *i.e. trace a unit circle in the complex plane as $\varphi$ ranges through the real numbers*

$\begin{split} 
z &= x + iy \newline
&= |z|(\cos \varphi + i \sin \varphi) \newline
&= re^{i\varphi} \newline
\end{split}$

$\begin{split} 
\bar{z} &= x - iy \newline
&= |z|(\cos \varphi - i \sin \varphi) \newline
&= re^{-i\varphi} \newline
\end{split}$

where
> $x = \Re(z)$ - the *real* part<br/>
> $y = \Im(z)$ - the *imaginary* part<br/>
> $r = |z| = \sqrt{x^2 + y^2}$ - the *magnitude* of z<br/>
> $\varphi = \arg z = \operatorname{atan2}(y, x)$ - the *argument* of z<br/>

Here $\varphi$ is also the angle

## Use formula to define the logarithm of a complex number

Both are valid for any complex number $a$ and $b$
> $a = e^{\ln a}$ <br/>
> $e^ae^b = e^{a+b}$ <br/>

Therefore
> $\begin{split}
z &= |z|e^{i\varphi} \newline
&= e^{\ln |z|}e^{i\varphi} \newline
&= e^{\ln|z| + i\varphi}
\end{split}$ <br/>
> *for any* $z \ne 0$

Complex Logarithm
> $\ln z = \ln |z| + i\varphi$ <br/>
> *this is a multi-valued function, as $\varphi$ is multi-valued*

## Relationship to trigonometry

Complex exponentials can simplify trigonometry, because they are easier to manipulate than their sinusoidal components. 
> One technique is simply to convert sinusoids into equivalent expressions in terms of exponentials. After the manipulations, the simplified result is still real-valued

### Sine & Cosine 

Cosine Definition
> $\begin{split}
\cos x &= \Re( e^{ix} ) \newline
&= \frac{e^{ix} + e^{-ix}}2
\end{split}$ <br/>

Sine Definition
> $\begin{split}
\sin x &= \Im( e^{ix} ) \newline
&= \frac{e^{ix} - e^{-ix}}{2i}
\end{split}$

#### Derivation

Use Euler's formula
> $e^{ix} = \cos x + i\sin x$ <br/>
> $e^{-ix} = \cos -x + i\sin -x = \cos x - i\sin x$

to solve for cos or sin
> $\begin{split}
e^{ix} &= \cos x + i\sin x \newline
&= \cos x + \cos x - e^{-ix} \newline
&= 2\cos x - e^{-ix} \newline
2\cos x &= e^{ix} + e^{-ix} \newline
\cos x &= \frac{ e^{ix} + e^{-ix} }2 \newline
\end{split}$

> $\begin{split}
e^{-ix} &= \cos x - i\sin x \newline
e^{ix} &= \cos x + i\sin x \newline 
e^{-ix} - e^{ix} &= \cos x - \cos x - 2i\sin x  \newline
e^{-ix} - e^{ix} &= - 2i\sin x \newline
2i\sin x &= e^{ix} - e^{-ix} \newline
\sin x &= \frac{ e^{ix} - e^{-ix} }{2i}
\end{split}$

### Complex arguments and Hyperbolic functions

$\cos iy = \frac{ e^{-y} + e^{y} }2 = \cosh y$

$\sin iy = \frac{ e^{-y} - e^{y} }{2i} = i\frac{ e^{y} - e^{-y} }{2} = i\sinh y$

### Simplification of Trigonometry

Complex exponentials can simplify trigonometry 

by replacing sinusoidal functions so that they can easily be manipulated.
> i.e. <br/>
> $\begin{split}
\cos{x}\cdot\cos{y} &= \frac{e^{ix} +e^{-ix}}2 \frac{e^{iy} +e^{-iy}}2 \newline
&= \frac12 \cdot \frac{ e^{i(x+y)} + e^{i(x-y)} + e^{i(-x+y)} + e^{i(-x-y)} }2 \newline
&= \frac12 \Big( \frac{ e^{i(x+y)} + e^{-i(x+y)} }2 + \frac{ e^{i(x-y)} + e^{-i(x-y)} }2 \Big) \newline
&= \frac12 \big( \cos(x+y) + \cos(x-y) \big) \newline
\end{split}$

or by representing sinusoids in terms of the real port of a complex expression
> i.e. <br/>
> $\begin{split}
\cos(nx) &= \Re(e^{inx}) \newline
&= \Re(e^{i(n-1)x} \cdot e^{ix}) \newline
&= \Re\big(e^{i(n-1)x} \cdot ( e^{ix} + e^{-ix} - e^{-ix} ) \big) \newline
&= \Re\big(e^{i(n-1)x} \cdot ( 2 \cos x - e^{-ix} ) \big) \newline
&= \Re(e^{i(n-1)x} \cdot 2 \cos x - e^{i(n-2)x} ) \newline
&= \cos [(n-1)x] \cdot [2cos(x)] - cos[(n-2)x]
\end{split}$ <br/>
>
> *This formula is used for recursive generation of $\cos(nx)$ for integer values of n and arbitrary x (in radians) as used in Phasor arithmetic

## History

### 1702 Bernoilli, Johann${}^1$

Noted that complex numbers
> $\frac{1}{1+x^2} = \frac12 \left(\frac{1}{1 - ix} + \frac{1}{1 + ix}\right)$

relate to natural logarithms
> $\msi \int \frac{dx}{1+ax} = \frac1a \ln(1 + ax) + C$

*Bernoulli's correspondence with Euler (who also knew the above equation) shows that Bernoulli did not fully understand. Euler also suggested that the complex logarithms can have infinitely many values.*

### 1714 Roger Cotes

Discovered that
> $ix = \ln(\cos x + i \sin x)$

*However, complex logarithms can have infinitely many values, due to trigonometric periodicity*

### 1740 Leonhard Euler

Used exponentials
> $\msi e^{ix} = \cos x + i \sin x$

### 1740 Leonhard Euler

Published the above formula

### 1790 Caspar Wessel

Described complex numbers as points in the complex pane
