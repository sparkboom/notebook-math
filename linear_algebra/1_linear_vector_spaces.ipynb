## Vector Space Definition

A vector (linear) space contains four things  $\{\F, \V, +, \times \}$
> $\textbf{F}$ field of scalars <br/>
> $\textbf{V}$ set of vectors <br/>
> $\textbf{+}$ Vector Addition *($\vec{v_1}$, $\vec{v_2} \in \V$ a new element $\vec{v_1} + \vec{v_2} \in \V$ )*<br/>
> $s\times$ Scalar Multiplication *($s \in \F$, $\vec{v} \in \V$ a new vector $s \sm \vec{v} \in \V$ )*<br/>


### Field of scalars

#### Field Definition
*A set of scalars ($\F$), where the following operation axioms hold.* <br/>

##### Addition ($+$) 
> $\begin{array} \\
\textsf{Axiom  LASA1} &\textsf{Closed} &a, b \in \F &\implies a + b \in \F \\
\textsf{Axiom  LASA2} &\textsf{Commutative} &a, b \in \F &\implies a + b = b + a \\
\textsf{Axiom  LASA3} &\textsf{Associative} &a, b, c \in \F &\implies (a + b) + c = a + (b + c) \\
\textsf{Axiom  LASA4} &\textsf{Identity (0)} &0, \forall a \in \F &\implies a+0 = 0 + a = a \\
\textsf{Axiom  LASA5} &\textsf{Inverse} &\forall a \in \F, \exists (-a) \in \F &\implies a+(-a) = (-a) + a = 0 \\
\end{array}$

##### Multiplication ($\cdot$)
> $\begin{array} \\
\textsf{Axiom  LASM1} &\textsf{Closed} &a, b \in \F &\implies a \cdot b \in \F \\
\textsf{Axiom  LASM2} &\textsf{Commutative} &a, b \in \F &\implies a \cdot b = b \cdot a \\
\textsf{Axiom  LASM3} &\textsf{Associative} &a, b, c \in \F &\implies (a \cdot b) \cdot c = a \cdot (b \cdot c) \\
\textsf{Axiom  LASM4} &\textsf{Identity (1)} &0, \forall a \in \F &\implies a \cdot 1 = 1 \cdot a = a \\
\textsf{Axiom  LASM5} &\textsf{Inverse} &\forall a \in \F, \exists (a^{-1}) \in \F &\implies a \cdot a^{-1} = a^{-1} \cdot a = 1 \\
\end{array}$

##### Distributivity
> $\begin{array} \\
\textsf{Axiom  LASD1} &\textsf{Left} &a, b, c \in F &\implies a \cdot(b+c) = a \cdot b + a \cdot c\\
\textsf{Axiom  LASD2} &\textsf{Right} &a, b, c \in F &\implies (a+b) \cdot c = a \cdot c + b \cdot c \\
\end{array}$

#### Field Examples
> - $\Q$ - Rational Numbers
> - $\R$ - Real Numbers 
> - $\C$ - Complex Numbers 
>
> *(Natural $\N$ and Integer $\Z$ numbers are not fields)*




### Vector Spaces

#### Vector Space Definition
*Vector Space $\V$ over a field $\F$ - denoted as $(\V, \F)$, are non-empty, are called **Vectors**, where the following operation axioms hold. <br/>

Addition ($+$)
> 1. Closed &#x2003;$u, v \in V \implies u + v \in V$
> 2. Commutative &#x2003;$u, v \in V \implies u + v = v + u$
> 3. Associative &#x2003;$u, v, w \in V \implies (u + v) + w = u + (v + w)$
> 4. Identity (denoted by 0) &#x2003;$\forall u \in V,\exists\; 0 \in V \implies u+0 = 0 + u = u$
> 5. Inverse $\forall u \in U, \exists (-u) \in V \implies  u+(-u) = (-u) + u = 0$


Scalar Multiplication ($a \cdot \vec{v}$)
> 1. Closed &#x2003;$u \in V, a \in F \implies a \cdot u = au \in V$
> 2. Field Distributive on Left &#x2003;$u,v \in V, a \in F \implies a(u + v) = au + av$
> 3. Field Distributive on Right &#x2003;$u \in V, a,b \in F \implies (a + b)u = au + bu$
> 4. Associative &#x2003;$u \in V, a,b \in F \implies ab(u) = a(bu)$
> 5. Unit Scalar Multiplication &#x2003;$u \in V, 1 \in F \implies 1u = u$

Negation
- $\vec{u},\vec{v} \in V\;\;\; \vec{u} - \vec{v}  \implies \vec{u} + (-\vec{v}) \implies \vec{u} + (-1)\cdot \vec{v}$

#### Vector Space Examples
- $(\V, \R),\, \R^n$ - Real Vector Space 
- $(\V, \C),\, \C^n$ - Complex Vector Space

### Theorems & Proofs
#### 1. $\vec{u} + \vec{v} = \vec{u} \implies \vec{v}=0 \in \V$
> $\begin{array} 
\em{Proof:} \\
\rlap{ \vec{u},\vec{v} \in \V} && \\ 
-\vec{u} +\vec{u} +\vec{v} &= -\vec{u} + \vec{u} &\implies \\
(-\vec{u} + \vec{u}) + \vec{v} &= 0 &\implies \\
0 + \vec{v} &= 0 &\implies \\
\therefore \vec{v} &= 0 &\blacksquare
\end{array}$

#### 2. $0 \cdot \vec{u} = 0 \in \V$
> $\begin{array} 
\em{Proof:} \\
\rlap{ \vec{u} \in \V}&& \\ 
0 \cdot \vec{u} &= (0 + 0)\vec{u} &\implies \\
&= 0 \cdot \vec{u} + 0 \cdot \vec{u} &\implies \\ 
\therefore 0 \cdot \vec{u} &\in \V &\blacksquare
\end{array}$

#### 3. $-\vec{u} \text{ is unique}$
> $\begin{array} 
\em{Proof:} \\
\rlap{ \vec{u}, \vec{v},\vec{w} \in \V}&& \\ 
\rlap{ \text{Assume } \vec{u}+\vec{v} = 0 \land \vec{u}+\vec{w} = 0}&& \\
\vec{v} &= \vec{v} + 0 &\implies \\
 &= \vec{v} + (\vec{u}+\vec{w}) &\implies \\
 &= (\vec{v} + \vec{u})+\vec{w} &\implies \\
 &= (\vec{u} + \vec{v})+\vec{w} &\implies \\
 &= 0 + \vec{w} &\implies \\
\therefore \vec{v} &= \vec{w} & \blacksquare  \\
\end{array}$ <br/>
>
> $-\vec{u}$ of any vector $u \in \V$ is unique <br/>
> or $\vec{u}_i,-\vec{u}_i \in \F | \vec{u}_i \mapsto -\vec{u}_i \land \vec{u}_j \mapsto -\vec{u}_i \iff i=j \toreview$ 


#### 4. $-u = (-1) \cdot u$
> $\begin{array} 
\em{Proof:} \\
\rlap{ \vec{u} \in \V}&& \\ 
0 &= 0 \cdot \vec{u} &\implies \\
 &= \big(1+(-1)\big) \vec{u} &\implies \\
 &= 1 \cdot \vec{u} + (-1)\vec{u} &\implies \\
\therefore 0 &= \vec{u} + (-1)\vec{u} &\blacksquare
\end{array}$ <br/>
>
> It follows that $(-1)u$ is a negative for $u$. <br/>
> The uniqueness of this negative vector now follows from Part 3.

### Poylnomial Space

#### Given
> $\displaystyle \PP_n = \Bigg\{ a_0 + a_1x + \dots a_{n-1}x^{n-1} = \sum\limits_{i=0}^{n-1} a_ix^i : a_i \in \F, x \in \R \Bigg\}$

#### Addition holds
> $\displaystyle 
a_i,b_i \in \F, x \in \R\\ 
\sum\limits_{i=0}^{n-1} \vec{a}_ix^i + \sum\limits_{i=0}^{n-1} \vec{b}_ix^i = \sum\limits_{i=0}^{n-1} (\vec{a}_i+\vec{b}_i)x^i$

#### Scalar Multiplication holds
> $\displaystyle  
a_i,b_i \in \F, x \in \R \\ 
c\sum\limits_{i=0}^{n-1} a_ix^i = \sum\limits_{i=0}^{n-1} (ca_i)x^i$


### Function Space

Let F(range) and X(domain) be fields, where X is a subset of F.
> $\mathbb{X} \subseteq \mathbb{F}$

The set of vectors V contains all the functions that map X to F, including continuous functions.
> $
(V,F) \subseteq \{ \forall f: \mathbb{X} \mapsto \mathbb{F} \} \\
(V,F) \subseteq (C[X], F) \\
$

Functions f and g are from the vector space V, x if from the field X (the domain), and x is from the field F(same as range)
> $f,g \in \mathbb{V}\\ x \in \mathbb{X}\\ c \in \mathbb{F}$

Additon and scalar multiplication of vector functions still hold
> $
f(x) + g(x) = (f+g)(x) \\
(cf)(x) = cf(x)
$

### Sequence Space

$\mathbb{S}$ is the set of all sequences with components from $\mathbb{F}$ a given field
> $\displaystyle \mathbb{S} = \Big\{\forall a | a=\{a_n\}_{n=1}^\infty, a_n \in \mathbb{F} \Big\}$

If a and b are sequences in $\mathbb{S}$, and c in $\mathbb{F}$
> $\displaystyle  a,b \in \mathbb{S} \\ c \in \mathbb{F}$

Addition and scalar multiplication of sequences still hold
> $a + b = \{a_n\} + \{b_n\} = \{a_n + b_n\}$ <br/>
> $ca_n = c\{a_n\} = \{ca_n\}$

($\mathbb{S}$, $\mathbb{F}$) is a vector space

### Homogenous Ordinary Linear Differential Equation Space

For the vector space (V, F)
> $\mathbb{F} = \mathbb{R}$ -  real constant coefficients <br/>
> $\mathbb{V}$ set of all solutions of the homogenous ordinary linear differential equation

Where
> $x \in \mathbb{R}$ <br/>

Homogenous ordinary linear differential equation
> $\displaystyle a_0\frac{d^ny}{dx^n} + a_1\frac{d^{n-1}y}{dx^{n-1}} + \dots + a_{n-1}\frac{dy}{dx} + a_ny = 0 \;|\; a_0 \ne 0$ <br/>

a.k.a
> $\displaystyle \sum\limits_{i=0}^n a_i\frac{d^{n-i}y}{dx^{n-i}} = 0, a_0 \ne 0$

Vector addition and scalar multiplication still hold

*(if the above is non homogeneous, then it won't hold)*

## Example - subspaces
Let $(\mathbb{V},\mathbb{F})$ and $(\mathbb{W},\mathbb{F})$ be vector spaces.
> $\mathbb{W} \subseteq \mathbb{V}$, then $(\mathbb{W}, \mathbb{F})$ is a subspace of $(\mathbb{V}, \mathbb{F})$ <br/>
> The smallest subspace $(\mathbb{W}, \mathbb{F})$ would be the zero vector <br/>
> The largest subspace $(\mathbb{W}, \mathbb{F})$ would be $(\mathbb{V}, \mathbb{F})$

Note that vectors with conditions may not be a subset <br/>
> If $\mathbb{W} = \left\{\begin{pmatrix}a_1\\a_2\\0\end{pmatrix} : a_1, a_2 \in \mathbb{R} \right\} $, and $\mathbb{V} = \left\{\begin{pmatrix}a_1\\a_2\\a_3\end{pmatrix} : a_1, a_2, a_3 \in \mathbb{R} \right\} $, then $(\mathbb{W}, \mathbb{F}) \subseteq (\mathbb{V}, \mathbb{F})$
<br/><br/>
> However, if $\mathbb{W} = \left\{\begin{pmatrix}a_1\\a_2\\a_3\end{pmatrix} : a_1>0, a_2>0,a_3>0 \right\} $, then $(\mathbb{W}, \mathbb{F}) \not\subseteq (\mathbb{V}, \mathbb{F})$

Powersets of lower degree are subsets <br/>
> $(\mathcal{P}_3, \mathbb{F}) \subseteq (\mathcal{P}_4, \mathbb{F})$ <br/>
> {all polynomials of degree exactly 2 over the field $\mathbb{F}$} $ \not\subseteq (\mathcal{P}_4, \mathbb{F}) $

Continuous functions in the set of all functions <br/>
> $(\mathbb{C[X], F}) \subseteq (\mathbb{V, F})$

### Theorem 1.2

> $(\V,\F), \W \subseteq \V, \mathbb{W} \ne \emptyset$<br/>
> $(\W, \F) \subseteq (\V, \F) \iff \forall \vec{u},\vec{v} \in \W, a \in \F;\;\;\; \exists a\vec{u} + \vec{v} \in \W$ <br/>

*Proof:* $\tolearn{}$ <br/>
> If $(\W, \F) \subseteq (\V, \F), \vec{u}, \vec{v} \in \W, a \in \F \implies a\vec{u} + \vec{v} \in \W.$ <br/>
> Conversely, $\W \ne \emptyset;\;\; \exists \vec{u} \in \W, a \in \F \implies a\vec{u} = a\vec{u} \in \W$ <br/>
> $\implies (-1)\vec{u} = -\vec{u} \in \W$ <br/>
> $\vec{u}, \vec{v} \in \W \implies 1\vec{u} + \vec{v} \in \W$ <br/>
> $\therefore (\W, \F) \subseteq (\V, \F)$ <br/>
> $\blacksquare$

> $(\W, \F) \subseteq (\V, \F) \iff \forall \vec{u},\vec{v} \in \W, \vec{u} + \vec{v} \in \W, a \in \F, a\vec{u} \in W$ <br/>

> $u_1,...,u_n \in \text{ some } (\V, \F),  c_1,...,c_n \in \F, $ <br/>
> $c_1\vec{u}_1 + ... + c_n\vec{u_n^n}$ is known as a *linear combination* of $u_i, i=1,...,n$ <br/>

By mathematical induction it follows that  <br/>
> $\vec{u} \in (\V, \F)$. <br/>

### Theorem 1.3

> $u^i \in (\V, \F), i = 1,...,n (\geq 1)$ <br/>
> $\W = \{c_1\vec{u_1} + ... + c_n\vec{u_n} : c_i \in \F, i=1,...,n \} \implies (\W, \F) \subseteq (\V, \F) \land \W \supset \{ u_i : i_1,...,i_n \}$ 

*Proof:* <br/>

each $u^i$ is a linear combination of the form <br/>
> $\displaystyle u_i = \sum\limits_{j=1}^{n} \delta_{i j}u^j $ <br/>

where $\delta_{i j}$ is the *Kronecker delta* defined by
> $\delta_{i j} = \left\{ \begin{array} \\ 0, & i \ne j \\ 1, & i = j \\ \end{array}\right.$


> $\forall i \in \W, \vec{v} = \sum\limits_{i=1}^n c_i u^i, \vec{w} = \sum\limits_{i=1}^n d_i u^i, a \in \F \implies $
> $a\vec{v} + \vec{w} = a\sum\limits_{i=1}^n c_i u^i + \sum\limits_{i=1}^n d_i u^i = \sum\limits_{i=1}^n (ac_i +d_i)u^i = $ <br/>
> $\sum\limits_{i=1}^n \alpha_i u^i, \alpha_i \in \F$ <br/>
> $\qeda$ Theorem 1.2

> $(\W, \F)$ is called the subspace *spanned* or *generated* by the vectors $u^i, i=1,...,n$, written as $Span\{u^1,...,u^n\}$ <br/>
> if $(\W, \F) = (\V, \F)$, then the set $\{u^1,...,u^n\}$ is called the *spanning set* for $(\V, \F)$
> each $\vec{u} \in \V$ can be expressed as a linear combination of vectors $\vec{u}^i, i=1,...,n$

> i.e.<br/>
> $2\begin{pmatrix}2\\1\\4\end{pmatrix} - 3\begin{pmatrix}1\\0\\2\end{pmatrix} + 5\begin{pmatrix}3\\2\\1\end{pmatrix} - \begin{pmatrix}4\\2\\0\end{pmatrix} = \begin{pmatrix}12\\10\\7\end{pmatrix}$ <br/>
> follows that $\begin{pmatrix}12\\10\\7\end{pmatrix} \in Span  \left\{\begin{pmatrix}2\\1\\4\end{pmatrix} , \begin{pmatrix}1\\0\\2\end{pmatrix} , \begin{pmatrix}3\\2\\1\end{pmatrix} , \begin{pmatrix}4\\2\\0\end{pmatrix} \right\}$ <br/>
> However, $\begin{pmatrix}1\\2\\3\end{pmatrix} \not\in Span \left\{ \begin{pmatrix}1\\0\\0\end{pmatrix} , \begin{pmatrix}1\\1\\0\end{pmatrix} \right\}$

Example 1.11
> $(\V, \F)$, the set $\{e_1,...,e_n\}$, where $e^i=\scriptscriptstyle\begin{pmatrix}0\\ \vdots\\ 0 \\ 1 \\ 0 \\ \vdots \\ 0 \end{pmatrix} \textstyle \in \V$ (1 at the ith place) is a spanning set.<br/>
> Similary for the vector space $(\PP_n, \F)$ in Example 1.3, and set $\{1, x, ...,x^{n-1}\}$ is a spanning set.
