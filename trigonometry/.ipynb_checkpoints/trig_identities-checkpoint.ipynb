# Trig Identities

### Trig Relations
> $\tan\theta = \frac{\sin\theta}{\cos\theta}$ <br/>
> $\cot\theta = \frac{1}{\tan\theta} = \frac{\cos\theta}{\sin\theta}$ <br/>
> $\sec\theta = \frac{1}{\cos\theta}$ <br/>
> $\csc\theta = \frac{1}{\sin\theta}$ <br/>

### Pythagorean Formula
> $\sin^2\theta + \cos^2\theta = 1$

### Trig Complements
> $\cos\theta = \sin(\frac{\pi}{2} - \theta)$ <br/>
> $\sin\theta = \cos(\frac{\pi}{2} - \theta)$ <br/>
> $\cot\theta = \tan(\frac{\pi}{2} - \theta)$ <br/>
> $\tan\theta = \cot(\frac{\pi}{2} - \theta)$ <br/>
> $\csc\theta = sec(\frac{\pi}{2} - \theta)$ <br/>
> $\sec\theta = csc(\frac{\pi}{2} - \theta)$

### Perodicity
> $\sin(\theta+2\pi) = \sin\theta$ <br/>
> $\cos(\theta+2\pi) = \cos\theta$ <br/>
> $\tan(\theta+2\pi) = \tan\theta$

### Negatives
> $\sin(-\theta) = -\sin\theta$ <br/>
> $\cos(-\theta) = \cos\theta$ <br/>
> $\tan(-\theta) = -\tan\theta$ 

### Ptolemy's Identities (sin/cos sum&difference formulas)
> $\sin(\alpha + \beta) = \sin\alpha\cos\beta + \cos\alpha\sin\beta$ <br/>
> $\cos(\alpha + \beta) = \cos\alpha\cos\beta - \sin\alpha\sin\beta$ <br/>
> $\sin(\alpha - \beta) = \sin\alpha\cos\beta - \cos\alpha\sin\beta$ <br/>
> $\cos(\alpha - \beta) = \cos\alpha\cos\beta + \sin\alpha\sin\beta$ <br/>
> $\sin 2\theta = 2 \sin\theta \cos\theta $

### Double Angle Formula
> $
\begin{split}
\cos 2\theta &= \cos^2\theta - \sin^2\theta \\
&= 2\cos^2\theta -1 \\
&= 1 - 2\sin^2\theta \\
\end{split}$

----

### Pythagorean Formula for tangents and secants
> $\sec^2\theta = 1 + \tan^2\theta$

### Trig in terms of their supplements
> $\sin(\pi - \theta) = \sin\theta$ <br/>
> $\cos(\pi - \theta) = -\cos\theta$ <br/>
> $\tan(\pi - \theta) = -\tan\theta$ 

### Sum, difference and double angle formulas for tangent
> $\tan 2\theta = \frac{2\tan\theta}{1 - \tan^2\theta}$ <br/>
> $\tan(\alpha + \beta) = \frac{\tan\alpha + \tan\beta}{1 - \tan\alpha\tan\beta}$ <br/>
> $\tan(\alpha - \beta) = \frac{\tan\alpha - \tan\beta}{1 + \tan\alpha\tan\beta}$

### Half angle formulas
> $\sin\frac\theta2 = \pm \sqrt{\frac{1 - \cos\theta}2}$ <br/>
> $\cos\frac\theta2 = \pm \sqrt{\frac{1 + \cos\theta}2}$ <br/>
> $\tan\frac\theta2 = \frac{\sin\theta}{1+\cos\theta} = \frac{1-\cos\theta}{\sin\theta}$

----

### Product-sum Identities
> $\sin\alpha + \sin\beta = 2\sin\frac{\alpha + \beta}2 \cos\frac{\alpha - \beta}2$ <br/>
> $\cos\alpha + \cos\beta = 2\cos\frac{\alpha + \beta}2 \cos\frac{\alpha - \beta}2$ <br/>
> $\sin\alpha - \sin\beta = 2\cos\frac{\alpha + \beta}2 \sin\frac{\alpha - \beta}2$ <br/>
> $\cos\alpha - \cos\beta = -2\cos\frac{\alpha + \beta}2 \sin\frac{\alpha - \beta}2$

### Product Identities
> $\sin\alpha \cos\beta = \frac{\sin(\alpha+\beta)+\sin(\alpha-\beta)}2$ <br/>
> $\cos\alpha \cos\beta = \frac{\cos(\alpha+\beta)+\cos(\alpha-\beta)}2$ <br/>
> $\sin\alpha \sin\beta = \frac{\cos(\alpha-\beta)-\cos(\alpha+\beta)}2$

### Triple Angle Formulas
> $\sin 3\theta = 3\sin\theta - 4\sin^3\theta$ <br/>
> $\cos 3\theta = 4\cos^3\theta - 3\cos\theta$ <br/>
> $\tan 3\theta = \frac{  3\tan\theta - \tan^3\theta  }{1 - 3\tan^2\theta}$

### Half angle formulas
> $\sin\theta = \frac{ 2\tan\frac\theta2 }{ 1 + \tan^2 \frac\theta2 }$ <br/>
> $\cos\theta = \frac{ 1- \tan^2 \frac\theta2 }{ 1 + \tan^2 \frac\theta2 }$ <br/>
> $\tan\theta = \frac{ 2\tan\frac\theta2 }{ 1 - \tan^2 \frac\theta2 }$ <br/>
