### Sums with a constant multiple

$\sum\limits_{n=s}^t \ia{C} \cdot \ib{f(n)} = \ia{C} \cdot \sum\limits_{n=s}^t \ib{f(n)}$
<br/>
$\ic{c}\id{y_s} + \ic{c}\id{y_{s+1}} + ... + \ic{c}\id{y_t} \;=\; \ic{c}(\id{y_s + y_{s+1} + ... + y_t}) $

### Sum of multiple functions

$\sum\limits_{n=s}^t \ia{f(n)} + \sum\limits_{n=s}^t \ib{g(n)} = \sum\limits_{n=s}^t \Big( \ia{f(n)} + \ib{g(n)} \Big)$
<br/>
$ (\ic{y_s + y_{s+1} + ... + y_t}) + (\id{w_s + w_{s+1} + ... + w_t}) \;=\; \Big( (\ic{y_s} + \id{w_s}) + (\ic{y_{s+1}} + \id{w_{s+1}}) + ... + (\ic{y_t} + \id{w_t}) \Big)$
<br/><br/>

$\sum\limits_{n=s}^t \ia{f(n)} - \sum\limits_{n=s}^t \ib{g(n)} = \sum\limits_{n=s}^t \Big( \ia{f(n)} - \ib{g(n)} \Big)$

$(\ic{y_s + y_{s+1} + ... + y_t}) - (\id{w_s + w_{s+1} + ... + w_t}) \;=\; \Big( (\ic{y_s} - \id{w_s}) + (\ic{y_{s+1}} - \id{w_{s+1}}) + ... + (\ic{y_t} - \id{w_t}) \Big)$


### Summing over indices

$\sum\limits_{n=s}^t \ib{f(n)} = \sum\limits_{n=s\ia{+p}}^{t\ia{+p}} \ib{f(n\ia{-p})} $

$\sum\limits_{n=s}^{\ia{j}} \ib{f(n)} + \sum\limits_{n=\ia{j+1}}^t \ib{f(n)} = \sum\limits_{n=s}^{t} \ib{f(n)} $


$\ia{ \sum\limits_{i=k_0}^{k_m} } \ib{ \sum\limits_{j=l_0}^{l_n} } a_{ij} = \ib{ \sum\limits_{j=l_0}^{l_n} } \ia{ \sum\limits_{i=k_0}^{k_m} } a_{ij} $
### Sum of staggered indicies

$ \sum\limits_{n=0}^t f(2n) + \sum\limits_{n=0}^t f(2n + 1) = \sum\limits_{n=0}^{2t+1} f(n) $

generalizes to 

$ \sum\limits_{n=0}^t \sum\limits_{i=0}^{z-1} f(z \cdot n + i) = \sum\limits_{n=0}^{z \cdot t + z-1} f(n)$

### Sum with the divisor function

$\sum\limits_{n \in A} f(n) = \sum\limits_{n \in \sigma(A)} f(n) $

*where $\sigma$ is the divisor function*

$c^{\big[ \sum_{n=s}^t f(n) \big]} = \prod\limits_{n=s}^t c^{f(n)} $

### Sum of indices

$ \sum\limits_{i=0}{n} i = \sum\limits_{i=1}{n} i = \frac{n(n-1)}{2}$

$ \sum\limits_{i=m}^n i = \frac{ n(n+1) }{2} - \frac{m(m-1)}{2} = \frac{ (n+1-m)(n + m) }{2} $

### Sum of powers

$ \sum\limits_{i=0}^n i^2 = \frac{n(n+1)(2n+1)}{6} = \frac{n^3}{3} + \frac{n^2}{2} + \frac{n}{6}$

$ \sum\limits_{i=0}^n i^3 = \Big( \frac{ n(n+1)}{2} \Big)^2 = \frac{n^4}{4} + \frac{n^3}{2} + \frac{n^2}{4} = \Bigg[ \sum\limits_{i=1}^n i \Bigg]^2  $

$ \sum\limits_{i=0}^n i^4 = \frac{ n(n + 1)(2n+1)(3n^2 + 3n -1) }{30} = \frac{n^5}{5} + \frac{n^4}{2} + \frac{n^3}{3} - \frac{n}{30} $

$ \sum\limits_{i=0}^n i^p = \frac{ (n+1)^{p+1} }{p+1} + \sum\limits_{k=1}{p} \frac{ B_k }{ p - k + 1} \begin{pmatrix}p\\k\end{pmatrix} (n+1)^{p-k+1} $

$ \Bigg( \sum\limits_{i=m}^n i \Bigg)^2 = \sum\limits_{i=m}^n \big(i^3 - im(m-1) \big) $

$ \sum\limits_{i=m}^n i^3 = \Bigg( \sum\limits_{i=m}^n i \Bigg)^2 + m(m-1) \sum\limits_{i=m}^n i $

### Sum over exponents

$ \sum\limits_{i=m}^n a^i = \frac{ a^m - a^n }{ 1 - a } $

$ \sum\limits_{i=0}^n a^i = \frac{ 1 - a^n }{ 1 - a } $

$ \sum\limits_{i=0}^n ia^i = \frac{ a - na^n + (n-1)a^{n+1} }{ (1 - a)^2 } $

$ \sum\limits_{i=0}^{n-1} i2^i = 2 + (n-2)2^n $

$ \sum\limits_{i=0}^{n-1} \frac{i}{2^i} = 2 - \frac{ n+1 }{ 2^{n-1}} $

### Sum of vectors?

$\sum\limits_{i=0}^n \begin{pmatrix}n\\i\end{pmatrix} = 2^n$

$\sum\limits_{i=1}^n i\begin{pmatrix}n\\i\end{pmatrix} = n2^{n-1}$

$\sum\limits_{i=0}^n \cdot \begin{pmatrix}n\\i\end{pmatrix} = \sum\limits_{i=0}^n {}_{n}P_i = \left \lfloor n! \cdot e \right \rfloor$

$\sum\limits_{i=0}^{n-1} \begin{pmatrix}i\\k\end{pmatrix} = \begin{pmatrix}n\\k+1\end{pmatrix} $

$\sum\limits_{i=0}^{n} \begin{pmatrix}n\\i\end{pmatrix}a^{n-i}b^i = (a+b)^n $

$\sum\limits_{i=0}^{n} i \cdot i! = (n+1)! - 1$

$\sum\limits_{i=0}^{n} {}_{i+k}P_{k+1} = \sum\limits_{i=0}^{n} \prod\limits_{j=0}^k (i + j) = \frac{ (n+k1)! }{ (n-1)!(k+2) }$

$\sum\limits_{i=0}^{n} \begin{pmatrix}m+i-1\\i\end{pmatrix} = \begin{pmatrix}m+n\\n\end{pmatrix}$
